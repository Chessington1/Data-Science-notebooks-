```python
import numpy as np
import matplotlib.pyplot as plt

# Data
donuts_per_person = {
    'Kwando': 6,
    'Jonathan': 8,
    'Jaco': 7,
    'Jason': 5,
    'Laura': 6,
    'Andrew': 8
}

# Extract data
donuts = donuts_per_person.values()
people = donuts_per_person.keys()

# Set positions for bars
y_pos = np.arange(len(donuts_per_person))

# Create bar chart
plt.bar(y_pos, donuts, alpha=0.5)
plt.xticks(y_pos, people)
plt.ylabel('Donuts')
plt.title('How many donuts each')

# Display chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to plot
labels = donuts_per_person.keys()
sizes = donuts_per_person.values()
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'grey', 'blue']

# Plot
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
total_donuts = {
    '31-01-2019': 40,
    '28-02-2019': 38,
    '31-03-2019': 30,
    '30-04-2019': 28,
    '31-05-2019': 25,
    '30-06-2019': 35
}

# Extract data
dates = list(total_donuts.keys())
x_ax = [date[3:5] for date in dates]  # Extracts month part
y_ax = list(total_donuts.values())

# Plot
plt.plot(x_ax, y_ax)
plt.xlabel('Month')
plt.ylabel('Number of Total Donuts Eaten')
plt.title('Line Plot showing the Number of Donuts Eaten over the First 6 Months of 2019')

# Display
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
donuts_per_age = {
    10: 25, 16: 22, 15: 20, 12: 21, 5: 30, 6: 31,
    10: 29, 2: 35, 4: 32, 8: 29, 7: 29, 2: 30,
    8: 31, 18: 22, 18: 24, 19: 25
}

# Extract data
x_ax = donuts_per_age.values()  # ages
y_ax = donuts_per_age.keys()    # number of donuts eaten

# Plot
plt.scatter(x_ax, y_ax, alpha=0.5)
plt.title('Scatter Plot Showing Number of Donuts Eaten per Age')
plt.xlabel('Age')
plt.ylabel('Number of Donuts Eaten')

# Display
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

# Load dataset
soccer = pd.read_csv("https://raw.githubusercontent.com/Explore-AI/Public-Data/master/Data/fundamentals/football_players.csv")

# Convert relevant columns to numerical values
cols = [
    'Age', 'Overall', 'Acceleration', 'Aggression', 'Agility', 'Balance', 
    'Ball control', 'Composure', 'Crossing', 'Curve', 'Dribbling', 'Finishing', 
    'Free kick accuracy', 'GK diving', 'GK handling', 'GK kicking', 
    'GK positioning', 'GK reflexes', 'Heading accuracy', 'Interceptions', 
    'Jumping', 'Long passing', 'Long shots', 'Marking', 'Penalties', 
    'Positioning', 'Reactions', 'Short passing', 'Shot power', 'Sliding tackle', 
    'Sprint speed', 'Stamina', 'Standing tackle', 'Strength', 'Vision', 'Volleys'
]



```

```python
soccer[cols] = soccer[cols].apply(pd.to_numeric, errors='coerce', axis=1)
```

```python
# Preview data
soccer.head()
```

```python
soccer = soccer.rename(columns= {'Preferred Positions' : 'Preferred_Positions'})
soccer.head()
```

```python
soccer.select_dtypes(include= ['object']).head()
```

```python
soccer['Age']= soccer['Age'].astype('float')
soccer['Age'].dtype
```

```python
sns.histplot(soccer['Overall'], bins=20, kde=True)
plt.title("Distribution of Player Overall Ratings")
plt.show()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 8))

sns.histplot(soccer['Age'], bins=5, kde=True, ax=ax[0])
sns.histplot(soccer['Age'], bins=5, kde=False, ax=ax[1])

ax[0].set_title('Distribution plot with Histogram for Ages')
ax[1].set_title('Histogram of Ages')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 3))
plt.xlabel('Age')
plt.hist(soccer['Age'], bins=5)
plt.show()
```

```python
j = sns.jointplot(data=soccer, x='Overall', y='Age', kind='reg')
plt.suptitle('Jointplot showing the Relationship between Overall and Age of Football Players', y=1.1)
plt.show()
```

```python
j = sns.jointplot(data=soccer, x='Overall', y='Age', kind='hex')
plt.suptitle('Jointplot showing the Relationship between Overall and Age of Football Players', y=1.1)
plt.show()
```

```python
young_players = soccer[(soccer['Age'] > 22) & (soccer['Age'] < 27)]  # filter ages
```

```python
plt.figure(figsize=(16,6))
sns.boxplot(data=young_players, x='Age', y='Overall')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.violinplot(data=young_players, x='Age', y='Overall')
plt.show()
```
