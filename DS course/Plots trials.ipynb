```python

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 6)
y = np.array([10, 15, 7, 12, 9])

plt.bar(x, y, color='skyblue')
plt.title("Simple Bar Chart Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
donuts_per_person = {
    'Kwando': 6,
    'Jonathan': 8,
    'Jaco': 7,
    'Jason': 5,
    'Laura': 6,
    'Andrew': 8
}

# Extract data
donuts = donuts_per_person.values()
people = donuts_per_person.keys()

# Set positions for bars
y_pos = np.arange(len(donuts_per_person))

# Create bar chart
plt.bar(y_pos, donuts, alpha=0.5)
plt.xticks(y_pos, people)
plt.ylabel('Donuts')
plt.title('How many donuts each')

# Display chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to plot
labels = donuts_per_person.keys()
sizes = donuts_per_person.values()
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'grey', 'blue']

# Plot
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
total_donuts = {
    '31-01-2019': 40,
    '28-02-2019': 38,
    '31-03-2019': 30,
    '30-04-2019': 28,
    '31-05-2019': 25,
    '30-06-2019': 35
}

# Extract data
dates = list(total_donuts.keys())
x_ax = [date[3:5] for date in dates]  # Extracts month part
y_ax = list(total_donuts.values())

# Plot
plt.plot(x_ax, y_ax)
plt.xlabel('Month')
plt.ylabel('Number of Total Donuts Eaten')
plt.title('Line Plot showing the Number of Donuts Eaten over the First 6 Months of 2019')

# Display
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
donuts_per_age = {
    10: 25, 16: 22, 15: 20, 12: 21, 5: 30, 6: 31,
    10: 29, 2: 35, 4: 32, 8: 29, 7: 29, 2: 30,
    8: 31, 18: 22, 18: 24, 19: 25
}

# Extract data
x_ax = donuts_per_age.values()  # ages
y_ax = donuts_per_age.keys()    # number of donuts eaten

# Plot
plt.scatter(x_ax, y_ax, alpha=0.5)
plt.title('Scatter Plot Showing Number of Donuts Eaten per Age')
plt.xlabel('Age')
plt.ylabel('Number of Donuts Eaten')

# Display
plt.show()
```
